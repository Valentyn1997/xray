```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
%pylab notebook
```

```python
img = cv2.imread('image.png',0)

edges = cv2.Canny(img,8,40)

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
edges = cv2.Canny(img,100, 250)

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
edges_high_thresh = cv2.Canny(img, 60, 120)
plt.subplot(122),plt.imshow(edges_high_thresh,cmap = 'gray')
plt.title('Edge Image 2'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
# Smoothing without removing edges.
gray_filtered = cv2.bilateralFilter(img, 5, 30, 30)

edges_filtered = cv2.Canny(gray_filtered, 10, 30)

plt.subplot(122),plt.imshow(edges_filtered,cmap = 'gray')
plt.title('Edge Image 3'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# compute the median of the single channel pixel intensities
v = np.median(img)
sigma = 0.05
# apply automatic Canny edge detection using the computed median
lower = int(max(0, (1.0 - sigma) * v))
upper = int(min(255, (1.0 + sigma) * v))
edged = cv2.Canny(img, lower, upper)
plt.subplot(122),plt.imshow(edged,cmap = 'gray')
plt.title('Edge Image 3'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
blurred = cv2.GaussianBlur(img, (3, 3), 0)
edges_blured = cv2.Canny(blurred, 10, 30)

plt.subplot(122),plt.imshow(edges_blured,cmap = 'gray')
plt.title('Edge Image 4'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img = cv2.imread('image.png',0)
ret,thresh = cv2.threshold(img,127,255,0)
contours,hierarchy = cv2.findContours(thresh, 1, 2)

cnt = contours[0]
M = cv2.moments(cnt)
rect = cv2.minAreaRect(cnt)
box = cv2.boxPoints(rect)
box = np.int0(box)
im = cv2.drawContours(img,[box],0,(0,0,255),2)
cv2.imshow("boxes", im)
```
