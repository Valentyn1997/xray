```python
import mlflow
import pandas as pd
import numpy as np
MLFLOW_TRACKING_URI = 'http://localhost:5001'
mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)
```

```python
cae = mlflow.search_runs(experiment_ids=10)
pd.merge(cae.groupby(by='tags.mlflow.parentRunId').agg(['mean', 'std']), cae, 
         left_on='tags.mlflow.parentRunId', right_on='run_id')
```
