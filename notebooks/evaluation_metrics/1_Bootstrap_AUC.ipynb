# Bootstrap ROC AUC

```python
import numpy as np
from scipy.stats import sem
from statistics import mean 
from sklearn.metrics import roc_auc_score
from sklearn.utils import resample
import matplotlib.pyplot as plt
```

```python
# set parameters
size = 1355 # size of data points
n_bootstrap = 1000 # size of bootstrap
random_seed = 42 # set random seed for reproducibility
```

## ROC AUC score 0.5

```python
# generate list with auc score 0.5
y_pred = list(range(0,size, 1))
y_true = []
for i in range(0, size):
    if(i % 2 == 0):
        y = 0
        y_true.append(y)
    if(i % 2 != 0):
        y = 1
        y_true.append(y)

# calculate auc score
score_original = roc_auc_score(y_true, y_pred)
print("Original ROC-AUC score: {:0.3f}".format(score_original))
```

## Bootstrapped ROC AUC


```python
# calculate bootstrap auc
bootstrapped_auc = []

y_pred = np.array(y_pred)
y_true = np.array(y_true)

for i in range(n_bootstrap):
    # bootstrap by sampling with replacement on the prediction indices
    indices = resample(y_pred, replace=True, n_samples=size+1, random_state=random_seed+i)
    score = roc_auc_score(y_true[indices], y_pred[indices])
    bootstrapped_auc.append(score)
    
    print("Bootstrap #{} ROC area: {:0.3f}".format(i + 1, score))
```

```python
plt.hist(bootstrapped_auc)
plt.title('Histogram bootstrapped ROC AUC scores')
plt.show()
```

## Calculate 0.95 Confidence Intervall

1. Calculate mean and ci based on mean
2. Sort bootstrapped AUC scores and print lower and upper ci

```python
mean_bootstrap = mean(bootstrapped_auc) # mean
sem_bootstrap = sem(bootstrapped_auc) # sem
upper_ci = mean_bootstrap + 1.96 * sem_bootstrap # upper ci = mean + 1.96(z score) * sem
lower_ci = mean_bootstrap - 1.96 * sem_bootstrap # lower ci = mean - 1.96(z score) * sem
print("Confidence interval 1 for the score: [{:0.3f}, {:0.3f}]".format(
    lower_ci, upper_ci))
```

```python
# sort bootstrapped auc scores
sorted_scores = np.array(bootstrapped_auc)
sorted_scores.sort()
# 
confidence_lower = sorted_scores[int(0.025 * len(sorted_scores))]
confidence_upper = sorted_scores[int(0.975 * len(sorted_scores))]
print("Confidence interval 2 for the score: [{:0.3f}, {:0.3f}]".format(
    confidence_lower, confidence_upper))
```
